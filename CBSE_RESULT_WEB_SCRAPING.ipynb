```python
from bs4 import BeautifulSoup
from openpyxl import Workbook

# Load the HTML result file (replace with your actual HTML string if needed)
with open("cbse_result_2025.html", "r", encoding="utf-8") as f:
    html = f.read()

soup = BeautifulSoup(html, 'html.parser')

# 1. Extract student info
student_info = {
    "Roll No": "",
    "Candidate Name": "",
    "Mother's Name": "",
    "Father's Name": "",
    "Date of Birth": ""
}

for td in soup.find_all("td"):
    text = td.get_text(strip=True)
    if "Roll No" in text:
        student_info["Roll No"] = td.find_next_sibling("td").get_text(strip=True)
    elif "Candidate Name" in text:
        student_info["Candidate Name"] = td.find_next_sibling("td").get_text(strip=True)
    elif "Mother's Name" in text:
        student_info["Mother's Name"] = td.find_next_sibling("td").get_text(strip=True)
    elif "Father's Name" in text:
        student_info["Father's Name"] = td.find_next_sibling("td").get_text(strip=True)
    elif "Date of Birth" in text:
        student_info["Date of Birth"] = td.find_next_sibling("td").get_text(strip=True)

# 2. Extract subjects and marks
subjects = []

for table in soup.find_all("table"):
    for row in table.find_all("tr"):
        cols = row.find_all("td")
        if len(cols) >= 6:
            sub_code = cols[0].get_text(strip=True)
            sub_name = cols[1].get_text(strip=True)
            theory = cols[2].get_text(strip=True)
            practical = cols[3].get_text(strip=True)
            total = cols[4].get_text(strip=True)
            grade = cols[5].get_text(strip=True)

            if sub_code.isdigit():
                short = sub_name.split()[0].upper()
                subjects.append({
                    "code": sub_code,
                    "short": short,
                    "total": total,
                    "grade": grade
                })

# 3. Extract final result (e.g., PASS/FAIL/COMPARTMENT)
final_result = ""
for td in soup.find_all("td"):
    text = td.get_text(strip=True)
    if text.upper().startswith("RESULT"):
        final_result = text.split(":")[-1].strip().upper()
        break

# 4. Create Excel
wb = Workbook()
ws = wb.active
ws.title = "CBSE Result"

# Prepare headers
headers = list(student_info.keys())
for sub in subjects:
    headers.append(f"{sub['short']}-{sub['code']}")
    headers.append("Grade")
headers.append("Result")

# Prepare data row
data = list(student_info.values())
for sub in subjects:
    data.append(sub['total'])
    data.append(sub['grade'])
data.append(final_result)

# Write to Excel
ws.append(headers)
ws.append(data)

# Save file
wb.save("CBSE_Result_22216627.xlsx")
print("Excel file 'CBSE_Result_22216627.xlsx' saved successfully.")
```
