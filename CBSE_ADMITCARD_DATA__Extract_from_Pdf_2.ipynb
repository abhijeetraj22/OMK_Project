```python
# Step 1: Install required libraries
!pip install pymupdf openpyxl

# Step 2: Upload the PDF file
from google.colab import files
uploaded = files.upload()

# Step 3: Import libraries
import fitz  # PyMuPDF
import re
import openpyxl
from openpyxl import Workbook

# Step 4: Open the uploaded PDF
pdf_file = next(iter(uploaded))
doc = fitz.open(pdf_file)

# Step 5: Combine text from all pages
full_text = ""
for page in doc:
    full_text += page.get_text()

# Step 6: Define regex patterns to extract required fields
patterns = {
    "Roll No.": r"Roll No\.?\s*[:\-]?\s*(\d+)",
    "Date of Birth": r"Date of Birth\s*[:\-]?\s*([\d/]{8,10})",
    "School No.": r"School No\.?\s*[:\-]?\s*(\d+)",
    "Admit Card ID": r"Admit Card ID\s*[:\-]?\s*(\w+)",
}

# Step 7: Extract values
extracted_data = {}
for label, pattern in patterns.items():
    match = re.search(pattern, full_text)
    if match:
        extracted_data[label] = match.group(1)
    else:
        extracted_data[label] = "Not Found"

# Step 8: Create an Excel file
wb = Workbook()
ws = wb.active
ws.title = "Extracted Data"

# Write headers and values
ws.append(["Field", "Value"])
for key, value in extracted_data.items():
    ws.append([key, value])

# Save the Excel file
excel_filename = "extracted_data.xlsx"
wb.save(excel_filename)

# Step 9: Download the Excel file
files.download(excel_filename)
```

```python
# Step 1: Install required libraries
!pip install pymupdf openpyxl

# Step 2: Upload the PDF file
from google.colab import files
uploaded = files.upload()

# Step 3: Import libraries
import fitz  # PyMuPDF
import re
import openpyxl
from openpyxl import Workbook

# Step 4: Open the uploaded PDF
pdf_file = next(iter(uploaded))
doc = fitz.open(pdf_file)

# Step 5: Combine text from all pages
full_text = ""
for page in doc:
    full_text += page.get_text()

# Step 6: Define regex patterns for multiple students
roll_nos = re.findall(r"Roll No\.?\s*[:\-]?\s*(\d+)", full_text)
dobs = re.findall(r"Date of Birth\s*[:\-]?\s*([\d/]{8,10})", full_text)
school_nos = re.findall(r"School No\.?\s*[:\-]?\s*(\d+)", full_text)
admit_ids = re.findall(r"Admit Card ID\s*[:\-]?\s*(\w+)", full_text)

# Step 7: Combine into rows (align by index)
num_students = min(len(roll_nos), len(dobs), len(school_nos), len(admit_ids))
data_rows = []
for i in range(num_students):
    data_rows.append([roll_nos[i], dobs[i], school_nos[i], admit_ids[i]])

# Step 8: Create Excel file with one row per student
wb = Workbook()
ws = wb.active
ws.title = "Students"

# Write headers
ws.append(["Roll No.", "Date of Birth", "School No.", "Admit Card ID"])

# Write student data
for row in data_rows:
    ws.append(row)

# Save the Excel file
excel_filename = "students_data.xlsx"
wb.save(excel_filename)

# Step 9: Download the Excel file
files.download(excel_filename)
```
