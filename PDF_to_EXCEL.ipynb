```python
import pytesseract
import camelot
from pdf2image import convert_from_path
from PIL import Image
import openpyxl
import io

pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

PDF_PATH = "/content/Grade-9 Science English Batch-6.pdf"
OUT_XLSX = "Grade-9Science English Batch-6.xlsx"

def hybrid_pdf_to_excel(pdf_path, out_xlsx):
    wb = openpyxl.Workbook()
    ws = wb.active
    ws.title = "PDF Data"
    row = 1

    # Try to detect tables using Camelot
    try:
        tables = camelot.read_pdf(pdf_path, pages="all")
    except:
        tables = []

    if len(tables) > 0:
        # Tables found → dump tables
        for t in tables:
            for r in t.df.values.tolist():
                for col, val in enumerate(r, start=1):
                    ws.cell(row=row, column=col).value = val
                row += 1
            row += 1

    else:
        # No tables → use OCR
        pages = convert_from_path(pdf_path)
        for p in pages:
            text = pytesseract.image_to_string(p, lang="eng")
            for line in text.split("\n"):
                if line.strip():
                    ws.cell(row=row, column=1).value = line.strip()
                    row += 1
            row += 1

    wb.save(out_xlsx)
    print("Saved:", out_xlsx)

hybrid_pdf_to_excel(PDF_PATH, OUT_XLSX)
```
